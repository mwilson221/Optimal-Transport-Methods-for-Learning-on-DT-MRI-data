```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import ot
import os

from scipy.stats import binom

from dtmrpy import DT_GMM
from dtmrpy.optimal_transport import wasserstein_type_distance

from time import time

device='cuda'
```

```python
from dtmrpy.optimal_transport import my_bures_log

import geomstats
from geomstats.geometry.spd_matrices import SPDMatrices, SPDBuresWassersteinMetric

space = SPDMatrices(3)
space.equip_with_metric(SPDBuresWassersteinMetric)

def covariance_projection(barycenter, mu, Pi, space):
    
        Sigma0=space.projection(barycenter.covariances_)
        Sigma0_update = np.zeros_like(Sigma0)
        # num_measures=len(measure_list)

#         for p, measure in enumerate(measure_list):

#             Pi=Pi_list[p]

        Sigma1=space.projection(mu.covariances_)

        for i in range(Sigma0.shape[0]):
            sigma0=Sigma0[i]
            ind = np.where(Pi[i]>0)[0]
            Sigma1_log = np.array([my_bures_log(sigma1,sigma0) for sigma1 in Sigma1[ind]])

            #Update Covariance
            temp=np.zeros((3,3))
            w_i=sum(Pi[i])
            for j in range(len(ind)):
                temp += Pi[i,ind[j]]*Sigma1_log[j]

            Sigma0_update[i] += space.metric.exp(temp/w_i,sigma0)

        return Sigma0_update
    
    
def log_map(base_point, mu):
    
    M = wasserstein_type_distance(barycenter,mu)[0]
    a = barycenter.weights_.flatten()
    b = (mu.weights_/sum(mu.weights_)).flatten()

    Pi = ot.emd(a,b,M)

    return DT_GMM(barycenter.weights_, (np.diag(1/a)@Pi@mu.means_), covariance_projection(barycenter, mu, Pi, space))
```

```python
df = pd.read_pickle("data_files/dtmri_dataframe_HCP.pkl")
df = df.dropna(subset=df.columns[2:],how='all').copy()
df.reset_index(inplace=True, drop=True)
df
```

```python
#folder to load barycenters from
barycenter_folder='C:\\Users\\micha\\Documents\\GitHub\\Optimal-Transport-Methods-for-Learning-on-DT-MRI-data\\data_files\\hcp\\'

#data_frame for stroring results
result_df = pd.DataFrame()

#Define Learner
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, StratifiedKFold, LeaveOneOut

svm = SVC()


#build models for each region
for barycenter_file in os.listdir(barycenter_folder):
    
    #load barycenter from file
    tract=barycenter_file[:-15]
    barycenter = np.load("data_files/hcp/"+tract+"_barycenter.npy", allow_pickle=True).item()

    #get barycentric_projections
    barycentric_projection_list = [log_map(barycenter,mu) for mu in df[tract] if type(mu)==DT_GMM]
    
    #get labels as integers
    y = [int(df['label'][i]=='M') for i, mu in enumerate(df[tract]) if type(mu)==DT_GMM]
    
    #for calculating p-value
    p = 1-np.mean(y)
    n = len(y)
    
    #get linear embedding - I think these are technically the barycentric projections
    X = np.zeros((200,3,len(barycentric_projection_list)))
    for i, mu in enumerate(barycentric_projection_list):
        X[:,:,i] = mu.means_-barycenter.means_
        
    #dimensionality reduction with tangent space PCA
    U,S,V = np.linalg.svd(np.cov(X.reshape(600,n)))

    #Project data into first 20 PC's
    X_star = V[:,:20].T@X.reshape(600,n)

    #Estimate accuracy with leave one out cross validation
    svm_score = cross_val_score(svm,X_star.T,y,cv=LeaveOneOut())
    
    #calculate p value
    # p = 1-np.mean(y)
    # n = len(y)
    p_val = 1-binom.cdf(np.mean(svm_score)*n, n, p)
    
    #store result
    result_df[tract]=[np.mean(svm_score), p_val]
    print(np.mean(svm_score))
```

```python
from sklearn.model_selection import StratifiedKFold, permutation_test_score
from sklearn.svm import SVC

clf = SVC()
cv = StratifiedKFold(2, shuffle=True, random_state=0)

score_iris, perm_scores_iris, pvalue_iris = permutation_test_score(
    clf, X_star.T, y, scoring="accuracy", cv=cv, n_permutations=1000
)

fig, ax = plt.subplots()

ax.hist(perm_scores_iris, bins=20, density=True)
ax.axvline(score_iris, ls="--", color="r")
score_label = f"Score on original\ndata: {score_iris:.2f}\n(p-value: {pvalue_iris:.3f})"
ax.text(0.7, 10, score_label, fontsize=12)
ax.set_xlabel("Accuracy score")
_ = ax.set_ylabel("Probability density")
```
